## TensorFlow Model

In this project, we will use TensorFlow to predict the value of NBA players, based on different features. 

```python
## first the general imports
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

## the training and test data related packages
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import make_column_transformer

## math related packages
from tensorflow import math

from tensorflow.keras.utils import plot_model

## testing to see if using TensorFlow GPU
if (tf.config.list_physical_devices('GPU')):
    print('Yes')
```

```python
## first loading the data
nba_data = pd.read_csv('C://Users//12145//Documents//GitHub//Python//data//nba_salary_stats.csv')
## checking the types and see if we have any null values
nba_data.info()
## other way could be to 
## nba_data.isna().sum()
## and we had nan 
## nba_data.dropna(inplace=True)
```

```python
## the next step is to remove any useless features
## and cleaning the data and reducing the memory usage
def data_cleaning(df, drop_cols = []):
    type_change_dict = {}
    if drop_cols:
        df = df.drop(drop_cols, axis=1)
    for col in df.columns:
        ## replacing all the non-alphabetic characters
        if df[col].dtype == 'object':
            df[col] = df[col].str.replace(r'\W', '', regex=True)
        ## changing the type when possible
        elif df[col].dtype == 'float64' and df[col].min() == df.astype({col:'float16'})[col].min():
            type_change_dict[col] = 'float16'
        elif df[col].dtype == 'float64' and df[col].min() == df.astype({col:'float32'})[col].min():
            type_change_dict[col] = 'float32'
        elif df[col].dtype == 'int64' and df[col].min() == df.astype({col:'int8'})[col].min():
            type_change_dict[col] = 'int8'
        elif df[col].dtype == 'int64' and df[col].min() == df.astype({col:'int16'})[col].min():
            type_change_dict[col] = 'int16'
    ## returning the new df with changed types
    return df.astype(type_change_dict)
    
cleaned_data = data_cleaning(nba_data, drop_cols=['first_name', 'last_name'])
## this simple function, cut the memory usage to less than half
cleaned_data.info()
```

```python
## the next step is to calculate the % of the outliers for each column
for col in cleaned_data.columns:
    if cleaned_data[col].dtype == 'object':
        continue
    q1 = cleaned_data[col].quantile(0.25)
    q3 = cleaned_data[col].quantile(0.75)
    percentage = cleaned_data[(cleaned_data[col] < (q1 - 1.5 * (q3 - q1))) |
                             (cleaned_data[col] > (q3 + 1.5 * (q3 - q1)))].shape[0] / cleaned_data.shape[0]
    print(f'{col} % of outlier: {round(percentage*100, 2)}%')
```

### TensorFlow Neural Network Regression
The next step is to start working on our actual model

```python
## first we will normalize our data
## and use one hot encoding for our non-numeric values
## we will use the MinMaxScaler and OneHotEncoder
## from sklearn pre-processing for this step
label_column = 'salary'
random_seed = 10
column_transformer = make_column_transformer((MinMaxScaler(), [x for x in cleaned_data.columns if cleaned_data[x].dtype!='object' and x != label_column]),
                                            (OneHotEncoder(handle_unknown='ignore'), [x for x in cleaned_data.columns if cleaned_data[x].dtype=='object' and x != label_column]))
## we also need to break our data into features and labels
X, Y = cleaned_data.drop(label_column, axis=1), cleaned_data[label_column]
## and then separate the training and test data sets
X_train, X_test, Y_tain, Y_test = train_test_split(X, Y, test_size=0.2, random_state=random_seed)

## and transforming our data
column_transformer.fit(X_train)
X_train_tf = column_transformer.transform(X_train)
X_test_tf = column_transformer.transform(X_test)
X_train_tf
```

And we're finally ready to work on our neural network for this project.

```python
## first we will set our random seed
## so we can reproduce the result later
tf.random.set_seed(random_seed)
## we will use a sequential model
## with 4 layers and reactified linear unit as our activation
## intead of the defaul linear model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(1),
])
## and then we want to compile our model
## using the mean absolute error as our loss
## and the Adam as our optimizer for our loss function
model.compile(loss = tf.keras.losses.mae,
             optimizer = tf.keras.optimizers.Adam(learning_rate=0.01),
             metrics=['mae'])
## and then use 100 epochs/iterations for our model
fit_data = model.fit(X_train_tf, Y_tain, epochs=100, verbose=0)
## and evaluate our model by test data
model_loss, model_mse = model.evaluate(X_test_tf, Y_test)
```

```python
## plotting the progress in the model
plt.figure(figsize = (8, 8))
sns.lineplot(data=fit_data.history)
```
