## Data Wrangling

In this project we will be using the [RT-IoT2022](https://archive.ics.uci.edu/dataset/942/rt-iot2022) dataset from UCI as our sample data.

```python
## first the imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## let's load the RT-IoT data
data = pd.read_csv('https://archive.ics.uci.edu/static/public/529/data.csv')
data.head()
```

```python
data.info()
```

```python
## which we can simply drop, since they're empty
## and caused by the poor data quality
def data_cleaner(df):
    drop_list = []
    ## lowercasing the column names
    df.columns = [x.replace(r'\s+', '_').lower() for x in df.columns]
    # drop_list = [x for x in df.columns if x.lower().startswith('unnamed')]
    for col in df.columns:
        if df[col].dtype == 'int64':
            df[col] = pd.to_numeric(df[col], downcast='unsigned')
        elif df[col].dtype == 'float64':
            df[col] = pd.to_numeric(df[col], downcast='float', errors='coerce')
        elif df[col].dtype == 'object' and df[col].nunique() == 2:
            
            df[col] = (pd.to_numeric(np.where(((df[col].str.lower()=='yes') |
                                              (df[col].str.lower() == 'female') |
                                              (df[col].str.lower() == 'positive')), 1, 0),
                                     downcast = 'unsigned'))
    return df
cleaned_data = data_cleaner(data.copy())
cleaned_data.info()
```

```python
cleaned_data.head()
```

```python
label = 'class'
nrows, ncols = 4, 4
c = 1
plt.figure(figsize = (16,16))
for col in cleaned_data.columns:
    if col == label:
        continue
    plt.subplot(nrows, ncols, c)
    if cleaned_data[col].nunique()==2:
        sns.countplot(x = col, data = cleaned_data, hue=label, palette='Blues')
    else:
        sns.histplot(x=col, data=cleaned_data, stat='density', kde=True)
    plt.title(col)
    c +=1
plt.tight_layout()
```

```python
facet = sns.FacetGrid(data=cleaned_data, row='obesity', col='class', hue = 'gender',palette='Greens')
facet.map(plt.hist, 'age');
```
